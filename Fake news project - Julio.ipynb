```python
def df_somatotalurl(df):
    """Receive a Dataframe with columns names 'related_tweet_content' and 'content' and returns a DataFrame with the sum of equal urls in each column and the total sum in all columns """
    #Separando links nos tweets de links retweets
    mask = df.related_tweet_content.notnull()
    dfrelatedtweet=df.copy()
    dfrelatedtweet=dfrelatedtweet[mask]
    
    mask2 = df.related_tweet_content.isnull()
    dfcontent=df.copy()
    dfcontent=dfcontent[mask2]
    
    # Criando uma coluna apenas com a url dos tweets/retweets
    regex = 'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    temp = [re.findall(regex, i) for i in dfrelatedtweet['related_tweet_content']]
    listaurl1=[]
    listaurl2=[]
    for lista in temp:
        try:
            listaurl1.append(lista[0])
        except:
            listaurl1.append(None)
        try:
            listaurl2.append(lista[1])
        except:
            listaurl2.append(None)
    dfrelatedtweet['url1']=listaurl1
    dfrelatedtweet['url2']=listaurl2
    dfcontent['content_URL'] = [''.join(re.findall(regex, i)) for i in dfcontent['content']]
    #Tirando as linhas sem url do DataFrame de url Content
    dfcontentsemvazio=dfcontent[dfcontent['content_URL'] != '']
    
    #Agrupando as urls por data(coluna que esta toda preenchida)
    dfcontentagrupado=pd.DataFrame(dfcontentsemvazio.groupby('content_URL')['date'].count())
    dfrelatedtweeturl1=pd.DataFrame(dfrelatedtweet.groupby('url1')['date'].count())
    dfrelatedtweeturl2=pd.DataFrame(dfrelatedtweet.groupby('url2')['date'].count())
    #Agrupando os Dataframes e somando as urls iguais
    dffoda=pd.merge(dfcontentagrupado, dfrelatedtweeturl1, how='outer', on=None, left_on=None, right_on=None,
         left_index=True, right_index=True, sort=False,
         suffixes=('_content', '_url1'), copy=True, indicator=False,
        )
    dffodao=pd.merge(dffoda, dfrelatedtweeturl2, how='outer', on=None, left_on=None, right_on=None,
          left_index=True, right_index=True, sort=False,
          suffixes=('_x', '_url2'), copy=True, indicator=False,
         )

    dffodao.fillna(0,inplace=True)
    dffodao['somaurl']=dffodao['date_content']+dffodao['date_url1']+dffodao['date']
    dffodao.sort_values('somaurl',ascending=False,inplace=True)
    dffodao.reset_index(level=0, inplace=True)
    dffodao=dffodao.rename(columns={'index':'URL','date_content':'sum_tweet_url','date_url1':'sum_rt_url1','date':'sum_rt_url2'})
    return dffodao
    
```
